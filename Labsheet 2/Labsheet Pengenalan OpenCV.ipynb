## Langkah-Langkah Pemrograman

### Import Library

```python
import cv2 as cv # Import library OpenCV dengan penamaan cv
```

### Membaca Gambar

```python
img = cv.imread('cat.png') # Isi dengan directory sesuai dengan gambar yang akan dibaca di pc kamu
```

```python
#Menampilkan gambar
cv.imshow('cat',img)
#Menunda windows terdestroy
cv.waitKey(0)
#Mendestroy windows
cv.destroyAllWindows()
```

### Menulis Ulang Gambar

```python
cv.imwrite('save_cat.png',img)
```

**Noted: Pada kode diatas untuk melihat dan menyimpan gambar, pastikan directory sesuai dengan tempat kalian menyimpan gambar tersebut**
